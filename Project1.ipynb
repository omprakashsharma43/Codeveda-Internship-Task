##### Task : Explotrary Data Analysis (EDA)

Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Generate synthetic Dataset

```python
np.random.seed(42)  # For reproducibility
n_samples = 500

data = pd.DataFrame({
    "CRIM": np.random.rand(n_samples) * 10,  # crime rate
    "ZN": np.random.randint(0, 100, n_samples),  # residential land zone
    "INDUS": np.random.uniform(0, 30, n_samples),  # industry proportion
    "CHAS": np.random.randint(0, 2, n_samples),  # Charles River dummy variable
    "NOX": np.random.uniform(0.3, 0.8, n_samples),  # nitric oxide concentration
    "RM": np.random.normal(6, 0.5, n_samples),  # avg rooms per dwelling
    "AGE": np.random.randint(20, 100, n_samples),  # age of houses
    "DIS": np.random.uniform(1, 12, n_samples),  # distance to employment centres
    "RAD": np.random.randint(1, 10, n_samples),  # accessibility to highways
    "TAX": np.random.randint(200, 700, n_samples),  # property tax
    "PTRATIO": np.random.uniform(12, 22, n_samples),  # pupil-teacher ratio
    "B": np.random.uniform(100, 400, n_samples),  # proportion of blacks
    "LSTAT": np.random.uniform(1, 40, n_samples),  # lower status population %
    "MEDV": np.random.uniform(5, 50, n_samples)   # median house value (target)
})

data
```

 Summary Statistics

```python
print("\nSummary Statistics:\n")
data.describe()
```

Visualizations

```python
# Histograms
data.hist(bins=20, figsize=(15, 10))
plt.suptitle("Histograms of Features", fontsize=16)
plt.show()

# Box plots
plt.figure(figsize=(15, 8))
sns.boxplot(data=data, orient="h")
plt.title("Box Plots of Features")
plt.show()

# Scatter plot example (RM vs MEDV)
plt.figure(figsize=(7, 5))
sns.scatterplot(x="RM", y="MEDV", data=data)
plt.title("Scatter Plot: Rooms vs House Value")
plt.show()
```

Correlation Matrix & Heatmap

```python
plt.figure(figsize=(12, 8))
corr = data.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

Insights (you can extend this section in a report)

```python
print("\nInsights:")
print("- RM (number of rooms) likely has a positive correlation with MEDV (house price).")
print("- LSTAT (lower status population %) likely has a negative correlation with MEDV.")
print("- TAX and RAD may show high correlation (multicollinearity).")
```
